```python
import pandas as pd
```

```python
df = pd.read_excel("df_2.xlsx")
df.info()
```

```python
df['nivel_educacion_madre_c'] = df['nivel_educacion_madre'].astype("category")
df['quintil_income_c'] = df['quintil_income'].astype("category")
df.info()
```

```python
df.drop(['nivel_educacion_madre','quintil_income'],axis=1,inplace=True)
df.head()
```

```python
df_ml = pd.get_dummies(df,prefix={'nivel_educacion_madre_c':'nem','quintil_income_c':'qin'},drop_first=True)
df_ml.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
cols_to_scale = ['Dist_cen','M_Dist_cen','Num_docentes','edad','p_matematica']
scaler = MinMaxScaler()
df_ml[cols_to_scale] = scaler.fit_transform(df_ml[cols_to_scale])
df_ml.head()
```

```python
dep_vars = ['p_matematica']
x = df_ml.drop(dep_vars,axis='columns')
y = df_ml[dep_vars]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```
