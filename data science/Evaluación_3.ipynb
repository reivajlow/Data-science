```python
data = pd.read_excel('data.xlsx')
data = pd.DataFrame(data)
data.head()
```

```python
#Separamos entre variables dependientes y independientes
import numpy as np
data.loc[data['Condition'] == 'M', 'Condition'] = 1.0
data.loc[data['Condition'] == 'B', 'Condition'] = 0.0
Dep_vars = ['Condition']
x = data.drop(Dep_vars, axis = 'columns')
y = data['Condition']
y = np.asarray(y).astype('float64')
data.info()
```

```python
#Entrenamiento y testeo
from sklearn.model_selection import train_test_split #Libreria para separar la data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20) #el random_state = semilla de rng
```

```python
#Librerias para los diferentes modelos
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
```

```python
model = Sequential()
#Input layer, input_dim, debe coincider con el numero de variables independientes
model.add(Dense(50, kernel_initializer = "normal", input_dim = 30, activation = "relu"))
#Hidden layer
#Cada capa con 60 neuronas
model.add(Dense(40, kernel_initializer = "normal", activation = "relu"))
model.add(Dense(40, kernel_initializer = "normal", activation = "relu"))
model.add(Dense(40, kernel_initializer = "normal", activation = "relu"))
#output layer
#Siempre es 1 neurona de respuesta
#Clasificacion usar sigmoid en activication
model.add(Dense(1, kernel_initializer = "normal", activation = "sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_split = 0.2, epochs = 200, verbose = False)
```

```python
# Confusion Matrix
from sklearn import metrics
pred = model.predict(X_test)
model.evaluate(X_train,y_train)
```
